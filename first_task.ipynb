```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv('intern.csv')
```

```python
df.head()
```

```python
missing_values = df.isna().sum()
```

```python
missing_values
```

```python
df.mean()
```

```python
df.fillna(df.mean(), inplace =True)
```

```python
missing_values = df.isna().sum()
```

```python
missing_values
```

```python
df.drop_duplicates(inplace = True)
```

```python
df
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
outlier_threshold = 1.5
df_no_outliers = df[~((df < (Q1 - outlier_threshold * IQR)) | (df > (Q3 + outlier_threshold * IQR))).any(axis=1)]
```

```python
df
```

```python
df.boxplot()
plt.title('Before Outlier Removal')
plt.show()
```

```python
df_no_outliers.boxplot()
plt.title('After Outlier Removal')
plt.show()
```
